```python
import cv2
import albumentations as A
import matplotlib.pyplot as plt

from custom import *
```

```python
def plot_images(title: str, origin, image, mask):
    fig, axes = plt.subplots(1, 3, figsize=(10, 3.5))

    axes[0].imshow(origin)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(image)
    axes[1].set_title('Image')
    axes[1].axis('off')

    axes[2].imshow(mask)
    axes[2].set_title('Mask')
    axes[2].axis('off')

    fig.suptitle(title, fontsize=16)

    plt.show()
```

```python
image_path = './data/image.png'
mask_path = './data/mask.png'

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
```

# Vertical Mirror Up

```python
vm_up = VerticalMirrorUp(p=1)
output = vm_up(image=image, mask=mask)
plot_images('VerticalMirrorUp', image, output['image'], output['mask'])
```

# Vertical Mirror Down

```python
vm_down = VerticalMirrorDown(p=1)
output = vm_down(image=image, mask=mask)

plot_images('VerticalMirrorDown', image, output['image'], output['mask'])
```

# Horizontal Mirror Up

```python
hm_up = HorizontalMirrorUp(p=1)
output = hm_up(image=image, mask=mask)

plot_images('HorizontalMirrorUp', image, output['image'], output['mask'])
```

# Horizontal Mirror Down

```python
hm_down = HorizontalMirrorDown(p=1)
output = hm_down(image=image, mask=mask)

plot_images('HorizontalMirrorDown', image, output['image'], output['mask'])
```

# Vertical Mirror Up + Horizontal Mirror Down

```python
transform = A.Compose([
    VerticalMirrorUp(p=1),
    HorizontalMirrorUp(p=1),
])

output = transform(image=image, mask=mask)

plot_images('Vertical Mirror Up + Horizontal Mirror Down', image, output['image'], output['mask'])
```
